```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Генерируем выборку случайно величины с равномерным распределением в диапазоне a...b

```python
a = 3
b = 6
obj_un = sts.uniform(a, b-a)
sample = obj_un.rvs(1000)
```

Строим гистограмму распределения по выборке и накладываем точное распределение

```python
x = np.linspace(2.9,7.5,100)
pdf = obj_un.pdf(x)

plt.hist(sample, normed='True', label ='experimental')
plt.plot(x, pdf, label='theoretical')
plt.legend(loc = 'upper right')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

# Оценка выборочных средних по выборкам объема n = 5, 10 ,15

```python
mean = np.zeros((1000,3))
n = np.array([5,10,15])

for i in range(3):
    for j in range(1000):
        mean[j,i] = np.sum( obj_un.rvs(n[i]) )/n[i]
```

Строим аппроксимацию выборочного среднего для n = 5,

тогда расчетные параметры нормального закона: $\mu=a+(b-a)/2$ и $\sigma^2=(b-a)^2/(12*n)$

```python
n = 5
mu = a + (b-a)/2.0
sigma = np.sqrt( (b-a)**2 / (12.0*n) )
x = np.linspace(3,6,100)

plt.hist(mean[:,0], normed='True', label ='experimental')
pdf = sts.norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, pdf, label='theoretical')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Аппроксимация выборочного среднего при n = 10

```python
n = 10
mu = a + (b-a)/2.0
sigma = np.sqrt( (b-a)**2 / (12.0*n) )
x = np.linspace(3,6,100)

plt.hist(mean[:,1], normed='True', label ='experimental')
pdf = sts.norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, pdf, label='theoretical')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Центральная предельная теорема позволяет с высокой точностью аппроксимировать выборочное среднее случайной величины с равномерным (и не только) законом распределения с помощью номального закона. Причем с ростом объема исходной выборки для оценки среденего точность аппроксимации увеличивается.
